```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/options risk management.csv')
```

```python
df.head()
```

```python
# Define trading strategy (Moving Average Crossover)
def moving_average_strategy(df, short_window, long_window):
    df['Signal'] = np.where(df['Underlying_Price'] > df['Underlying_Price'].rolling(window=short_window).mean(), 1,
                           np.where(df['Underlying_Price'] < df['Underlying_Price'].rolling(window=long_window).mean(), -1, 0))
    df['Position'] = df['Signal'].diff()
    return df
```

```python
# Backtest strategy
df = moving_average_strategy(df, 10, 30)
```

```python
# Calculate strategy returns
df['Strategy_Returns'] = df['Underlying_Price'].pct_change() * df['Position'].shift(1)

# Evaluate strategy performance
strategy_returns = df['Strategy_Returns'].dropna()
sharpe_ratio = strategy_returns.mean() / strategy_returns.std() * np.sqrt(252)
sortino_ratio = strategy_returns.mean() / strategy_returns[strategy_returns < 0].std() * np.sqrt(252)
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Sortino Ratio: {sortino_ratio:.2f}")
```

```python
# Plot strategy performance
plt.figure(figsize=(12, 6))
plt.plot(df['Underlying_Price'], label='Underlying Price')
plt.plot(df['Underlying_Price'] * df['Position'], label='Strategy')
plt.legend()
plt.title('Strategy Performance')
plt.show()
```

```python
# Plot strategy returns
plt.figure(figsize=(12, 6))
plt.plot(strategy_returns.cumsum(), label='Strategy Returns')
plt.legend()
plt.title('Cumulative Strategy Returns')
plt.show()
```

```python
# Group data by Option Type and Risk Level
option_type_risk_level = df.groupby(['Option_Type', 'Risk_Level']).size().reset_index(name='Count')

# Pivot data for plotting
pivot_df = option_type_risk_level.pivot(index='Option_Type', columns='Risk_Level', values='Count')

# Plot Option Type wise Risk Level
pivot_df.plot(kind='bar', figsize=(10, 6), title='Option Type wise Risk Level')
plt.xlabel('Option Type')
plt.ylabel('Count')
plt.legend(title='Risk Level')
plt.show()
```

```python
# Group data by Option Type and calculate average Strike Price
option_type_avg_strike_price = df.groupby('Option_Type')['Strike_Price'].mean().reset_index()

# Plot Option Type wise Average Strike Price
plt.figure(figsize=(10, 6))
plt.bar(option_type_avg_strike_price['Option_Type'], option_type_avg_strike_price['Strike_Price'])
plt.title('Option Type wise Average Strike Price')
plt.xlabel('Option Type')
plt.ylabel('Average Strike Price')
plt.show()

```
